## Exemplo 2: Avaliação de um aluno 
Construa um programa que leia as notas da P1, P2 e P3 e avalie se aluno está aprovado, recuperação ou reprovado. A P3 possui peso 2.
- Aprovado Nf > 6
- Recuperação 5 <= Nf <= 6 
- Reprovado Nf < 6
- Nf = (P1 + P2 + 2*P3)/4

```python
notas = []
P1 = float(input('Digite a nota P1: '))
P2 = float(input('Digite a nota P2: '))
P3 = float(input('Digite a nota P3: '))
Nf = (P1 + P2 + 2*P3)/4
if Nf >=6:
  print('Aprovado!')
elif Nf <= 6 and Nf >=5:
  print('O aluno está de recuperação')
else:
  print('Aluno reprovou')
print(f'A média final: {Nf}')
```

##Exemplo 3: Resolução de Equação de segundo grau. 
Construa um programa para resolução de equação do segundo grau utilizando a fórmula de Bhaskara, onde indica ao usuário as seguintes mensagens.
- delta > 0, há solução real, neste caso, as raízes da equação são: 
- delta = 0, há solução real, neste caso, as raíz da equação é:
- delta < 0, não há solução real.
\
y = x²-2x +1\
\
663800148e0118776884fe6e4d70712f.jpg
\


```python
import sys
a = float(input('Digite o coeficiente "a": '))
if a == 0:
  print('Equação de segundo grau inexistente')
  sys.exit()
b = float(input('Digite o coeficiente "b": '))
c = float(input('Digite o coeficiente "c": '))
delta = (b**2)-4*a*c
if delta > 0:
  x1 = (-b+(delta**0.5))/2*a
  x2 = (-b+(delta**0.5))/2*a
  print(f'delta > 0, há solução real, neste caso, as raízes da equação são: {x1} e {x2}')
elif delta == 0:
  x1 = -b/2*a
  print(f'delta = 0, há solução real, neste caso, as raíz da equação é: {x1}')
else:
  print('delta < 0, não há solução real')
```

##DESAFIO 1: Escala de pH
Construa um programa que recebe o valor da concentração de H3O+ e determine o seu valor de pH e se a suposta solução é ácida, básica ou neutra. Lembre-se que:\
pH = -log[H3O+]


- Utilize as bibliotecas math ou numpy, caso achar necessário na resolução do exercício, pesquise na documentação utilizando os links oferecidos.

```python
import math
concentracao = float(input('Insira a concentração de H3O+:  '))
pH = -1*math.log(concentracao, 10)
if pH>7:
  print(f'Solução é básica com o pH de {pH:.1E}')
elif pH == 7:
  print(f'Solução neutra com pH de {pH:.1E}' )
else: 
  print(f'Solução ácida com pH de {pH:.1E}')
```

##DESAFIO II: Cálculo de pH parte II 
Construa um programa leia os valores da constante ácida e concetração molar, então calcule o pH considerando as seguintes condições:\

- para ácidos fortes, em concentrações acima de 1E10-6 mol/l. 

- para ácidos fracos, em concentrações acima de 1E10-6 mol/l. 

- para ácidos fortes, em concetntrações abaixo de 1E10-6 mol/l (considerar efeito de autoprotólise).

CLassifique:
- Ka > 1 para ácidos fortes
- Ka < 1 para ácidos fracos

Dados do Problema\

```python
import math  #importação da biblioteca math
ka = float(input('Constante ácida (Ka): '))  #leitura de dados utilzando o comando input
conc = float(input('Concentração, em mol/L: '))  #leitura de dados utilzando o comando input
if ka > 1:  #ka>1 engloba ácidos fortes
  if conc >= 1e-6:  #1e-6 é maior que 0, é desconsiderado efeitos de protólise
    pH = -math.log(conc,10)  #resolução do desafio 1
    print(f'Ácido forte sem efeitos de autoprotólise, pH = {pH}')  #resultado para ocasião 1
  else:  #1e-6 é menor que 0, portanto, considera-se efeitos de protólise
    a = 1
    b = -conc
    c = -1e-14
    if a == 0: 
      print('Equação de segundo grau inexistente')
    else:
      delta = (b**2)-(4*a*c)
      if delta > 0:
        x1 = (-b+(delta**0.5))/(2*a)
        x2 = (-b-(delta**0.5))/(2*a)
        if x1>=0:  #esta estrutura de decisão seleciona somente valores positivos
          pH = -1*math.log(x1,10)
          print(f'Ácido forte com efeitos de autoprotólise, pH = {pH}')  #resultado para ocasião 2
        elif x2>=0:
          pH = -1*math.log(x2,10)
          print(f'Ácido forte com efeitos de autoprotólise, pH = {pH}')  #resultado para ocasião 2
        else: 
          print('Não há resolução')
      elif delta == 0:
        x1 = -b/2*a
        if x1>=0: #esta estrutura de decisão seleciona somente valores positivos
          pH = -1*math.log(x1,10)
          print(f'Ácido forte com efeitos de autoprotólise3, pH = {pH}')  #resultado para ocasião 2
        else:
          print('Não há resolução')
else:  #ka<1 engloba ácidos fracos
  h30 = (conc*ka)**0.5
  pH = -1*math.log(h30,10)
  print(f'Ácido fraco sem efeitos de autoprotólise, pH = {pH}')  #resultado para ocasião 3
```
